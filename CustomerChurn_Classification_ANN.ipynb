#### Crossvalidated model

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    churn_classif = Sequential()
    churn_classif.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = 12))
    churn_classif.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'relu'))
    churn_classif.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    churn_classif.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return churn_classif
```

```python
churn_model = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 50)
```

```python
accuracies = cross_val_score(churn_model, X_train, y_train, cv = 10, n_jobs = -1)
```

```python
accuracies
```

```python
accuracies.mean()
```

#### Fix for potential overfitted ANN 

```python
# In the event of overfitting we could use Dropout regularization
```

```python
from keras.layers import Dropout
```

```python
def build_classifier():
    churn_classif = Sequential()
    churn_classif.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = 12))
    churn_classif.add(Dropout(p = 0.2))
    churn_classif.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'relu'))
    churn_classif.add(Dropout(p = 0.2))
    churn_classif.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    churn_classif.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return churn_classif
```

```python
churn_model = KerasClassifier(build_fn = build_classifier, batch_size = 10, epochs = 50)
accuracies = cross_val_score(churn_model, X_train, y_train, cv = 10, n_jobs = -1)
```

#### Tuning the model

```python
from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier():
    churn_classif = Sequential()
    churn_classif.add(Dense(units = 8, kernel_initializer = 'uniform', activation = 'relu', input_dim = 12))
    churn_classif.add(Dense(units = 5, kernel_initializer = 'uniform', activation = 'relu'))
    churn_classif.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    churn_classif.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return churn_classif
```

```python
tuned_model = KerasClassifier(build_fn = build_classifier)

params = {'batch_size':[10, 25, 32],
          'epochs':[50, 100, 500],
#           'optimizer': ['adam', 'rmsprop']
         }

tuned_model_grid = GridSearchCV(tuned_model, params, cv = 10, scoring = 'accuracy')
tuned_model_grid.fit(X_train, y_train)
```

```python
print("Best Params:",tuned_model_grid.best_params_)
print("Best Cross Val Score:",tuned_model_grid.best_score_)
```
